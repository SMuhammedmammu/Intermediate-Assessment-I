````python
df = pd.read_csv('C:/Users/USER/Dropbox/PC/Downloads/diabetes_model.csv')
df
````

````python
# Univariate Analysis: Histogram
df['Glucose'].hist(bins=25, color='skyblue', edgecolor='black')
plt.title('Histogram of Glucose')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()

# Univariate Analysis: Box Plot
sns.boxplot(x=df['Glucose'], color='orange')
plt.title('Box Plot of Glucose')
plt.show()
````

````python
# Bivariant Analysis: Scatter Plot: Insulin vs BMI
sns.scatterplot(x='BMI', y='Insulin', data=df, color='purple')
plt.title('Insulin vs BMI')
plt.xlabel('BMI')
plt.ylabel('Insulin')
plt.show()

# Bivariant Analysis Correlation Matrix
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
````

````python
# Pair Plot
sns.pairplot(df, hue='Outcome', palette='husl')
plt.show()
````

````python
# 3D Plot (requires mpl_toolkits.mplot3d)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['BloodPressure'], df['Glucose'], df['Age'], c=df['Outcome'], cmap='viridis')
ax.set_xlabel('BloodPressure')
ax.set_ylabel('Glucose')
ax.set_zlabel('Age')
plt.title('3D Scatter Plot: BloodPressure, Glucose, Age')
plt.show()
````
